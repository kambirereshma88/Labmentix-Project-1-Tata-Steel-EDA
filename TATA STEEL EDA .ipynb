## chart4 : Correlation Heatmap

```python
#step 1: Identify non numeric columns
non_numeric_cols = df.select_dtypes(exclude=[np.number]).columns
print("Non-Numeric Columns:", non_numeric_cols)
```

```python
# Step 2: Convert numeric columns stored as strings
for col in non_numeric_cols:
    try:
        df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert if possible
    except:
        print(f"Could not convert {col}")
```

```python

# Step 3: Keep only numeric columns
df_numeric = df.select_dtypes(include=[np.number])  
```

```python
# Step 4: Check if numeric columns exist before plotting
if not df_numeric.empty:
    plt.figure(figsize=(10, 6))
    sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title("Correlation Heatmap")
    plt.show()
else:
    print("No numeric columns available for correlation.")
```

#### Why did you pick the specific chart?

A correlation heatmap helps visualize relationships between numerical variables, guiding feature selection for machine learning models.

What insights are found from the chart?

Strong correlations indicate which features influence each other, while weak correlations suggest independent variables.

Will the gained insights help create a positive business impact?

Yes, understanding correlations enhances predictive modeling and decision-making. However, multicollinearity issues should be managed to prevent model instability.

##  Chart 7: Histogram of Tool Wear

The presence of outliers suggests some tools may wear out much faster than expected, possibly due to poor maintenance, improper usage, or material defects.

Will the gained insights help create a positive business impact?
